#### Проверка адекватности в задачах регрессии (критерии согласия)

**Определение**:

Пусть $x_1, x_2, .., x_n$ - независимые наблюдения случайно величины с функцией распределения $F(x)$, которая нам не известна. Предположим, что нам нужно проверить гипотезу:

$H_0: F(x) =F_0(X),$

где $F_0(x)$ - некоторая заданная функция распределения, которая может быть непрерывной или дискретной. Задача проверки гипотезы называется задачей *проверки согласия*, а критерии, используемые для этого, *критериями согласия*.

**В регрессионном анализе критерии согласия позволяют оценивать значимость регрессионных моделей.**

Вывод о значимости гипотезы делают на основе значения вероятности(p) соответствия проверяемой гипотезы тестовой выборке (статистическим данным).

Для принятия гипотезы необходимо, чтобы $p > 0,5$ (если иное не оговаривается используемым критерием согласия).

### Статистические критерии

Наша конечная цель — определить статистическую значимость наших результатов. И статистическая значимость построена на этих 3 простых идеях:

- Проверка гипотезы
- Нормальное распределение
- P-значение (p-value)

Проверка гипотез используется для проверки обоснованности утверждения (нулевой гипотезы), сделанного в отношении совокупности с использованием выборочных данных. Альтернативная гипотеза — это та, в которую вы бы поверили, если бы нулевая гипотеза оказалась неверной.

Чтобы узнать, является ли утверждение обоснованным или нет, мы будем использовать p-значение для взвешивания силы доказательств, чтобы увидеть, является ли оно статистически значимым. Если доказательства подтверждают альтернативную гипотезу, то мы отвергнем нулевую гипотезу и примем альтернативную гипотезу.

##### Критерий Пирсона (критерий $χ^2$)

Статистический критерии согласия Пирсона используют следующие нулевую и альтернативную гипотезы:

$H_0$ : (нулевая гипотеза) Переменная следует за гипотетическим распределением.

$H_1$ : (альтернативная гипотеза) Переменная не подчиняется предполагаемому распределению.

Мы используем следующую формулу для расчета статистики критерия хи-квадрат, $\chi^2$ ,

$\chi^2 = \sum_{i=1}^k \frac {(y_i - y_i^*)^2}{y_i^*}$, где $k$ - количество проверяемых значений.

Если p-value, соответствующее тестовой статистике $\chi^2$ с $n-1$ степенями свободы (где $n$ — количество категорий), меньше выбранного вами уровня значимости (обычно выбираются 0,10, 0,05 и 0,01), то вы можете отвергнуть нулевую гипотезу.

##### Критерий Фишера

Обычно он используется в качестве альтернативы критерию независимости хи-квадрат.

Точный тест Фишера использует следующие нулевые и альтернативные гипотезы:

$H_0$ : (нулевая гипотеза) Переменные независимы.

$H_1$ : (альтернативная гипотеза) Переменные не являются независимыми.

В общем виде F-критерий Фишера рассчитывается по следующей формуле:

$F = \frac{S_{факт}^2}{S_{ост}^2}$, где $S_{факт}^2$ - факторная дисперсия, $S_{ост}^2$ - остаточная дисперсия. 
Соответствующие дисперсии определяются по следующим формулам:

$S_{факт} = \frac{\sum(y^* - \bar{y})^2}{m}$, $S_{ост} = \frac{\sum(y - y^*)^2}{n - m - 1}$

В приведенных формулах $n$ – это число наблюдений, $m$ – число параметров при переменной $x$ (то есть количество факторов в модели регрессии).

Интерпретация F-критерия Фишера для уравнения регрессии в целом следующая: в том случае, когда фактическая величина F-критерия Фишера больше табличного показателя, то уравнение регрессии в целом является статистически значимым.

##### Критерий Стьюдента (t-тест)

t-критерий для парных выборок всегда использует следующую нулевую гипотезу:

$H_0$ : $µ_1 = µ_2$ (средние значения двух выборок равны)

Альтернативная гипотеза может быть двусторонней, левосторонней или правосторонней:

$H_1$ (двухсторонняя): $μ_1 ≠ μ_2$ (средние значения двух выборок не равны)

$H_1$ (левосторонняя): $μ_1 < μ_2$ (среднее значение для выборки 1 меньше среднего значения для выборки 2)

$H_1$ (правосторонняя): $μ_1 > μ_2$ (среднее значение для выборки 1 больше, чем среднее для выборки 2)

Мы используем следующую формулу для расчета тестовой статистики t:

$t =\frac{\bar{x}_1 - \bar{x}_2}{\sqrt{m_1^2 + m_2^2}} $,

где $\bar{x}_{i}$ - среднее значение, 

$m_i = \frac{sd}{\sqrt{n}}$, $sd$ - стандартное отклонение, $n$ - количество измерений.

Если p-value, соответствующее тестовой статистике t с (n-1) степенями свободы, меньше выбранного вами уровня значимости (обычно выбираются 0,10, 0,05 и 0,01), вы можете отклонить нулевую гипотезу.

### Метрики

В случае если планируется делать обобщения и использовать модель на других данных, то необходимы множественная проверка результатов модели на различных данных и использование оценки точности, такие как:

1. **MSE (Mean Square Error)** – одна из самых популярных метрик в задаче регрессии. 

$MSE(y^{true}, y^{pred}) = \frac1N\sum_{i=1}^{N} (y_i - f(x_i))^2$

Иногда для того, чтобы показатель эффективности MSE имел размерность исходных данных, из него извлекают квадратный корень и получают показатель эффективности **RMSE**.

$RMSE = \sqrt{MSE}$

MSE неограничен сверху, и может быть нелегко понять, насколько "хорошим" или "плохим" является то или иное его значение. Чтобы появились какие-то ориентиры, делают следующее:

- Берут наилучшее константное предсказание с точки зрения MSE — среднее арифметическое меток . При этом чтобы не было подглядывания в test, среднее нужно вычислять по обучающей выборке

- Рассматривают в качестве показателя ошибки:

$R^2 = 1 - \frac{\sum_{i=1}^{N} (y_i - f(x_i))^2}{\sum_{i=1}^{N} (y_i - \bar{y})^2}.$

У идеального решающего правила $R^2$  равен 1, у наилучшего константного предсказания он равен 0 на обучающей выборке. Можно заметить, что $R^2$
показывает, какая доля дисперсии таргетов (знаменатель) объяснена моделью.

2. **MAE(mean absolute error)**

Использовать RMSE для сравнения моделей на выборках с большим количеством выбросов может быть неудобно. В таких случаях прибегают к также знакомой вам в качестве функции потери метрике MAE (mean absolute error):

$MAE(y^{true}, y^{pred}) = \frac{1}{N}\sum_{i=1}^{N} \left|y_i - f(x_i)\right|$

#### Метрики, учитывающие относительные ошибки

В метрике MAPE (mean absolute percentage error) в знаменатель помещают целевое значение:

$MAPE(y^{true}, y^{pred}) = \frac{1}{N} \sum_{i=1}^{N} \frac{ \left|y_i - f(x_i)\right|}{\left|y_i\right|}$

```python
# подготовим данные так же, как на семинаре 5
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data_path = "C:/Users/Анна/Desktop/ML занятия/HSE ML/HSE ML/DataSets"
# Стоимость и объем продаж акций компании Google (работа с временными рядами, может быть решена задача прогнозирования)
data1 = pd.read_csv(data_path + "/DataSet1_3.csv")
# Выберем данные
n1 = 101
n2 = 300
data = data1[n1:n2].copy()
data['Date'] = pd.to_datetime(data['Date'], format="%Y-%m-%d")
data['Seconds'] = ((data['Date'] - data['Date'].iloc[0]).dt.total_seconds())/(60*60*24) + 1
data = data.rename(columns = {'Close':'y', 'Seconds':'x'})
data = data.drop(columns=['Date', 'High', 'Low', 'Volume', 'Name'])
data['lag'] = data1['Close'].iloc[n1-1:n2].shift(1).copy()

data.head()
```

```python
# разбиваем данные на тренировочную и тестовую выборку
X = data.drop(columns=['y'])
y = data['y']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

### МНК



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(np.array(x_train), np.array(y_train))
res_mnk = model.predict(np.array(x_test))
```

#### Визуальный анализ


```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].plot(x_test['x'], res_mnk, label='МНК')
axs[0].scatter(x_test['x'], y_test, color='red', label='Данные')
axs[0].set_title('Визуальная оценка результатов')
axs[0].legend()
axs[0].set_xlabel('X')

axs[1].boxplot([ y_test, res_mnk])
axs[1].set_title('Boxplot для анализа результатов')
axs[1].set_xticklabels(['Данные', 'Прогноз'])
plt.show()
```

По графиками мы можем сравнить характеристики разброса значений: медианные значения практически совпадают, границы квартилей чуть сдвинуты (*что это значит? подумайте!*), "границы усов" — границы статистически значимой выборки, тоже несколько сдвинуты, разброс значений по прогнозу чуть больше, чем реальные значения. #### Статистическая оценка
Для выполнения анализа с использованием статистических критериев согласия в Python, вы можете воспользоваться библиотекой scipy.

```python
import scipy.stats as stats

# критерий хи-квадрат

def cs(y_pred, y_test):
    return stats.chisquare(y_pred, np.sum(y_pred)/np.sum(y_test) * y_test)

print("Результат критерий Пирсона(хи квадрат)")
print(cs(res_mnk, y_test))
```

```python
# тест Фишера

# Посчитаем дисперсии
variance1 = np.var(y_test, ddof=1)
variance2 = np.var(res_mnk, ddof=1)
 
# Посчитаем F
f_value = variance1 / variance2
 
# Почитаем степени свободы у выборок
df1 = len(y_test) - 1
df2 = len(res_mnk) - 1
 
# Посчитаем p-value
p_value = stats.f.cdf(f_value, df1, df2)
 
print("Результат критерий Фишера\n")
print("F-statistic:", f_value)
print("p-value:", p_value)
```

```python
# t-тест

print("Результат t-теста\n")
print(stats.ttest_rel(y_test, res_mnk)) #для зависимых величин
# print(stats.ttest_ind(y_test, res_mnk)) #для двух независимых величин
```

### Метрики

```python
# для практики, давайте посчитаем руками

def regression_error_metrics(model, x_train, y_train, X_test, y_test, model_name=''):
    model.fit(x_train, y_train)
    y_pred = model.predict(X_test)
    n = len(y_test)
    
    MSE = (1/n) * np.sum((y_test - y_pred)**2)
    RMSE = MSE**0.5
    R2 = 1 - (np.sum((y_test - y_pred)**2))/(np.sum((y_test - np.mean(y_test))**2))
    MAE = (1/n) * np.sum(abs(y_test - y_pred))
    MAPE = (1/n) *  np.sum(abs((y_test - y_pred)/y_test))
        
    model_error_metrics = {
        'MSE': MSE,
        'RMSE': RMSE,
        'R2': R2,
        'MAE': MAE,
        'MAPE': MAPE}
    
    result = pd.DataFrame({
        'MSE': MSE,
        'RMSE': RMSE,
        'R2': R2,
        'MAE': MAE,
        'MAPE': "{:.3%}".format(MAPE)},
        index=[model_name])        
        
    return model_error_metrics, result

_, result = regression_error_metrics(model, x_train, y_train, x_test, y_test, model_name='Least Squares')

result
```
